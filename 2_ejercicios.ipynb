```python
# ejercicio 6
a = np.array([1, 3, 5])
b = np.array([2, 3, 4])

mayor = a > b
mayor_igual = a >= b
menor = a < b
menor_igual = a <= b

print("a > b :", mayor)
print("a >= b :", mayor_igual)
print("a < b :", menor)
print("a <= b :", menor_igual)
```
